```python
def get_news_data(Article_Name):
    print(f"Loading {Article_Name}")
    NEWS_COLUMNS = ['DATE', 'SECTION', 'COMPANY', 'TITLE', 'MAIN', 'LINK']
    data = pd.read_csv(f".././news_data/{Article_Name}",
                      encoding='cp949', names =  NEWS_COLUMNS)
    print(f"Load Complete! Total {data.shape[0]} contents are available.")
    return data
```

```python
# 안봐도됨
def tokenizer(doc, pos_method):
    return ['/'.join(t) for t in pos_method.pos(doc, norm=True, stem=True)]
```
